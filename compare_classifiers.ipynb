```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
rnn_acc = np.array([0.74629629, 0.74444443, 0.74259257, 0.74074072, 0.74629629,
       0.72962964, 0.73703706, 0.73148149, 0.72777778, 0.73148149])
```

```python
model_names = ['Baseline', 'Random Forest', 'RNN']
accuracies = [0.6, 0.6, np.mean(rnn_acc)]
stds = [0,0, np.std(rnn_acc)]

# Create a bar plot
sns.set_style("whitegrid")
plt.figure(figsize=(12,4))
sns.barplot(x=model_names, y=accuracies, errorbar=('ci', 95), yerr=stds, palette='rocket')
plt.ylim((.5,0.8))
# Add labels and titles
plt.xlabel('Classifier')
plt.ylabel('Accuracy (on test set)')
plt.title('Comparison of Model Accuracy on Test Set')
plt.show()
```
